```r
df <- read.csv("/content/WorldCupMatches.csv")
attendance <- na.omit(df$Attendance)

n <- length(attendance)
k <- ceiling(1+3.322*log10(n))
amplitude <- max(attendance) - min(attendance)
ampclasse <- amplitude / k

print(ampclasse)

classes <- seq(min(attendance), max(attendance), by = ampclasse)

frequencias <- cut(attendance, breaks = classes, include.lowest = TRUE, right = FALSE)

tabela_frequencia <- table(frequencias)

hist(attendance,
  breaks = seq(min(attendance), max(attendance), by = ampclasse),
  main = "Historiograma com Distribuição de Torcida no Jogos",
  xlab = "Quantidade de Torcedores",
  ylab = "Quantidade de Jogos",
  col = "lightgreen",
  border = "black"
)
```
