```r
# Importação dos dados
df <- read.csv2("/content/dataframe_1ap.csv")
```

```r
# COLUNA 1
# Separar as colunas de faturamento
df_faturamento_col1 <- df$fat_pdv1

### Distribuição de frequência fat_pdv1
n_col1 <- length(df_faturamento_col1)
k_col1 <- ceiling(1 + 3.322 * log10(n_col1))
amplitude <- max(df_faturamento_col1) - min(df_faturamento_col1)
ampclasse <- amplitude / k_col1
classes <- seq(min(df_faturamento_col1), max(df_faturamento_col1), by = ampclasse)

print(paste("A coluna fat_pdv1 tem a diferença de classe de:", ampclasse))
for (i in seq_len(length(classes) - 1)) {
  cat(
    sprintf(
      "range %d: de %.2f a %.2f\n",
      i,
      classes[i],
      classes[i + 1]
    )
  )
}

### Tabela de frequências
frequencias_col1 <- cut(df_faturamento_col1, breaks = classes, include.lowest = TRUE, right = FALSE)
tabela_frequencias_col1 <- table(frequencias_col1)
tabela_print_col1 <- format(tabela_frequencias_col1, scientific = FALSE)
print(tabela_print_col1)

### Histograma
# Histograma Coluna fat_pdv1
histograma_col1 = hist(df_faturamento_col1, breaks = classes,
          main = "Histograma com Distribuição de Frequências pdv_1",
          xlab = "faturamento",
          ylab = "frequência",
          col = "lightblue",
          border = "black")

for(i in 1:length(histograma_col1$counts)) {
  # Colocando o texto no topo de cada barra
  text(
    histograma_col1$mids[i],                          # Posição no eixo X (mídia de cada classe)
    histograma_col1$counts[i],                        # Posição no eixo Y (frequência)
    labels = histograma_col1$counts[i],               # Rótulo com o valor da frequência
    pos = 3,                                     # Posiciona o texto acima da barra
    cex = 0.8,                                   # Ajusta o tamanho do texto
    col = "black"                                # Cor do texto
  )
}
```
