```r
salarios <- c(1900,1900,2200,3500,4000,5000,2800,1900,1900,1400)
n <- length(salarios) # Quantidade de elementos na amostra
k = ceiling(1 + 3.322 * log10(10)) # ceiling é como se fosse um "Round Up"
amplitude <- max(salarios) - min(salarios) # Amplitude total
ampclasse <- amplitude / k # Amplitude de classe: variação entre os valores de cada classes
classes <- seq(min(salarios), max(salarios), by = ampclasse)
print(classes)
```

```r
frequencias <- cut(salarios, breaks = classes, include.lowest = TRUE, right = FALSE)

tabela_frequencias <- table(frequencias)
print(tabela_frequencias)
```

```r
hist(salarios,
breaks = seq(min(salarios), max(salarios), by = ampclasse),
main = 'Histograma com Distribuição de Frequências',
xlab = "Valores",
ylab = "Frequência",
col = 'lightblue',
border = 'black'
)
```

```r
media <- mean(salarios)
print(media)

frequencia <- table(salarios)
moda <- as.numeric(names(frequencia)[which.max(frequencia)])
print(moda)

mediana <- median(salarios)
print(mediana)

variancia <- var(salarios)
print(variancia)

desvio_padrao <- sd(salarios)
print(desvio_padrao)
```

```r
boxplot(salarios, main = 'Boxplot', ylab = "Valores", col = 'lightgreen')
```
